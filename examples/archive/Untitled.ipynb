```python
import pyqsp
import pyqsp.completion
import pyqsp.angle_sequence
import pyqsp.phases
import numpy as np
from matplotlib import pyplot as plt
from pyqsp.gadgets import *
```

```python
n = 25
```

```python
Q = pyqsp.poly.PolyExtraction().generate(degree=n)
P = pyqsp.completion._qp_completion(Q)
one = P * np.polynomial.Polynomial(P.coef.conj()) + np.polynomial.Polynomial([1, 0, -1]) * Q * np.polynomial.Polynomial(Q.coef.conj())
```

```python
X = np.linspace(-1, 1, 100)

plt.plot(X, one(X))
plt.plot(X, P(X).real)
plt.plot(X, P(X).imag)
plt.plot(X, Q(X).real)
plt.plot(X, (lambda x : 1/np.sqrt(1-x**2))(X))
```

```python
phi = pyqsp.phases.ExtractionSequence().generate(n)
fn = lambda x : compute_mqsp_unitary([W(x)], phi, [0 for _ in range(len(phi)-1)])[0][1]
```

```python
plt.plot(np.linspace(-1, 1, 100), [np.abs(fn(x)) for x in np.linspace(-1, 1, 100)])
plt.plot(np.linspace(-1, 1, 100), [1 for x in np.linspace(-1, 1, 100)])
```
